**Passeio Aleatório e Probabilidade**

```python
"""
APOSTILA: PASSEIO ALEATÓRIO E PROBABILIDADE
Curso: Ciências da Computação
Disciplina: Introdução a Processos Estocásticos
"""

# ============================================================
# 1. O que é um Passeio Aleatório?
# ============================================================
"""
O passeio aleatório (random walk) é um modelo que descreve
um movimento em que, a cada passo, a posição pode variar
aleatoriamente.

Exemplo clássico: uma pessoa em uma linha reta que,
a cada instante, tem 50% de chance de andar um passo à
frente (+1) ou um passo atrás (-1).

Esse processo é governado pela PROBABILIDADE.
"""

# ============================================================
# 2. Representação com Variáveis Aleatórias
# ============================================================
"""
Cada passo pode ser modelado por uma variável aleatória X_i:

X_i = +1 (com probabilidade 0,5)
X_i = -1 (com probabilidade 0,5)

A posição após n passos é a soma acumulada:

S_n = X_1 + X_2 + ... + X_n
"""

import numpy as np
import matplotlib.pyplot as plt

# Definindo o número de passos
n = 140

# Sorteando os passos (+1 ou -1)
steps = np.random.choice([-1, 1], size=n)

# Posição acumulada (soma dos passos)
position = np.cumsum(steps)

# Plotando o passeio aleatório
plt.plot(position)
plt.title("Passeio Aleatório 1D")
plt.xlabel("Tempo (passos)")
plt.ylabel("Posição")
plt.grid(True)
plt.show()

# ============================================================
# 3. Relação com a Probabilidade
# ============================================================
"""
- Valor Esperado (média) de cada passo:
  E[X_i] = (+1)(0,5) + (-1)(0,5) = 0
  => A posição média tende a ficar próxima de 0.

- Variância:
  Cresce proporcionalmente ao número de passos.
  => Quanto mais passos, maior a dispersão.

- Lei dos Grandes Números:
  Se repetirmos muitas vezes, a média tende a estabilizar.
"""

# Simulação de várias trajetórias para observar a dispersão
plt.figure(figsize=(10,6))
for i in range(10):  # 10 simulações diferentes
    steps = np.random.choice([-1, 1], size=n)
    position = np.cumsum(steps)
    plt.plot(position, alpha=0.7)

plt.title("Múltiplas Simulações do Passeio Aleatório")
plt.xlabel("Tempo (passos)")
plt.ylabel("Posição")
plt.grid(True)
plt.show()

# ============================================================
# 4. Aplicações
# ============================================================
"""
O passeio aleatório aparece em diversas áreas:

- Redes de Computadores: tráfego de pacotes.
- Finanças: modelagem de preços de ações.
- Ciência de Dados: simulações de Monte Carlo.
- Física: movimento browniano de partículas.
- IA: exploração em algoritmos de aprendizado por reforço.

✅ Resumo: O passeio aleatório é simples, mas revela como
a probabilidade governa fenômenos dinâmicos e incertos.
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000  # número de passos
steps = np.random.choice([-1, 1], size=n)
position = np.cumsum(steps)

plt.plot(position)
plt.title("Passeio Aleatório 1D")
plt.xlabel("Tempo (passos)")
plt.ylabel("Posição")
plt.grid(True)
plt.show()
```
